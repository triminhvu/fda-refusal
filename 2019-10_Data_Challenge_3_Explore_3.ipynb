# FDA Import Refusal - Exploration 3

In this exploration, I group import refusals by country. When plotting import refusals over time for countries we see that from 2002-2018, import refusals from Canada decreases while import refusals from China steadily increases. We can also see import refusals spike in certain country such as the United Kingdom in 2008.

```python
import pandas as pd

import holoviews as hv
hv.extension('bokeh')

#for choropleth map
#pip install descartes
import geopandas as gpd

import matplotlib.pyplot as plt
```

First, we load the unskewed import refusal dataset. This dataset has the entries for surgical trays from Convertors De Mexico removed.

```python
df_refusal = pd.read_pickle('fda_refusal_2002_2019_unskewed.pkl')
```

We group number of import refusal by country.

```python
df_country = df_refusal.groupby(['REFUSAL_DATE', 'ISO_CNTRY_CODE'])['LGL_NAME'].count()
df_country = df_country.reset_index()
df_country.columns = ['Refusal Date', 'iso_a2', 'Number of Refusals']
```

FDA dataset use ISO 3116 alpha 2 as the country code. We want to merge the country name and the ISO 3116 alpha 3 code for further graphing and analysis

```python
country_codes = pd.read_csv('FDA_Country_Codes.csv')
country_codes.columns = ['iso_a2', 'Country']
iso_country_codes = pd.read_csv('ISO_Country_Codes.csv', encoding='cp1252')
iso_country_codes = iso_country_codes.dropna()
iso_country_codes = iso_country_codes.drop(columns = ['Country_Name', 'Official_Name', 'Sovereignty', 
                                              'ISO_3166-1_numeric', 'ISO_3166-2', 'Internet_ccTLD'])
iso_country_codes.columns = ['iso_a2', 'iso_a3']
country_codes = country_codes.merge(iso_country_codes)
pd.to_pickle(country_codes,'Country_Codes.pkl')
country_codes.head()
```

```python
df_country = df_country.merge(country_codes)
```

```python
df_country['Year'] = df_country['Refusal Date'].dt.year
df_country['Year-Month'] = df_country['Refusal Date'].dt.strftime('%Y-%m')
df_country['Year-Month'] = pd.to_datetime(df_country['Year-Month'])
df_country.head()
```

We group the monthly refusals by country. We see that on average country have 20 refusal per month with the standard deviation of 43. If we only look at country with large number of refusal, we see that India and China dominate.

```python
df_country_YM = df_country.groupby(['Year-Month', 'Country', 'iso_a3'])['Number of Refusals'].sum()
df_country_YM = df_country_YM.reset_index()
df_country_YM.describe()
```

We want to see the top country with refusal.

```python
df_country_YM_large = df_country_YM[df_country_YM['Number of Refusals'] > 106]
hv.Bars(
    data = df_country_YM_large,
    kdims=['Year-Month','Country'],
    vdims=['Number of Refusals'],
).opts(
    height=600,
    width=900,
    padding=0.1,
    cmap = 'Category20',
    legend_position = 'right',
    tools=['hover'],
    stacked = True
)
```

We see the same thing if we group the annual refusals by country.

```python
df_country_Y = df_country.groupby(['Year', 'Country', 'iso_a3'])['Number of Refusals'].sum()
df_country_Y = df_country_Y.reset_index()
df_country_Y.describe()
```

```python
hv.Bars(
    data = df_country_Y,
    kdims=['Year','Country'],
    vdims=['Number of Refusals'],
).opts(
    height=600,
    width=900,
    padding=0.1,
    cmap = 'Category10',
    legend_position = 'right',
    tools=['hover'],
    stacked = True
)
```

```python
df_country_Y_large = df_country_Y[df_country_Y['Number of Refusals'] > 490]
hv.Bars(
    data = df_country_Y_large,
    kdims=['Year','Country'],
    vdims=['Number of Refusals'],
).opts(
    height=600,
    width=900,
    padding=0.1,
    cmap = 'Category20',
    legend_position = 'right',
    tools=['hover'],
    stacked = True
)
```

We can observe the change overtime using choropleth map.

```python
df_by_year = df_country_Y.loc[df_country_Y['Year']==2002, ['iso_a3','Number of Refusals']]
df_by_year.columns = ['iso_a3', 'Number of Refusals 2002']
col_names = []
for year in range(2002,2019,1):
    df = df_country_Y.loc[df_country_Y['Year']==year, ['iso_a3','Number of Refusals']]
    col_name = 'Number of Refusals ' + str(year)
    col_names.append(col_name)
    df.columns = ['iso_a3', col_name]
    df_by_year = df_by_year.merge(df)
df_by_year.head()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world[world.name!="Antarctica"]
world = world.merge(df_by_year)
```

```python
for col in col_names:
    ax = world.plot(column = col,
               figsize = (10, 10),
               legend=True,
               legend_kwds={'label': "Number of Refusals",'orientation': "horizontal"}
    )
    ax.set_title(col)
```

We can observe import refusals over time for certain country. In this case we have Canada.

```python
hv.Curve(
    data = df_country_YM[df_country_YM['Country'] == 'Canada'],
    kdims=['Year-Month'],
    vdims=['Number of Refusals'],
).opts(
    title = 'Canada',
    height=400,
    width=600,
    padding=0.1,
    tools=['hover'],
)
```

We can also try to see if there is seasonality to import refusal from certain country. In this case Canada.

```python
df_canada = df_country_YM[df_country_YM['Country'] == 'Canada']
df_canada['Month'] = df_canada['Year-Month'].dt.month
df_canada_M_ave = df_canada.groupby('Month')['Number of Refusals'].mean()
```

```python
points = hv.Points(
    data = df_canada,
    kdims=['Month','Number of Refusals'],
    vdims=['Number of Refusals'],
).opts(
    title="Canada Import Refusal by month",
    height=400,
    width=600,
    padding=0.01,
    xticks = 12,
    show_grid=True,
    tools=['hover'],
)

line = hv.Curve(
    data = df_canada_M_ave,
    kdims=['Month'],
    vdims=['Number of Refusals']
)
points * line
```
